```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("youtube_dataset.csv")

category_views = df.groupby("Category")["Views_in_millions"].sum()

plt.figure(figsize=(10,6))
plt.bar(category_views.index, category_views.values,
        color="red", edgecolor="black", width=0.5)

plt.title("Total Views per Video Category", fontsize=14, fontweight="bold")
plt.xlabel("Video Category", fontsize=12)
plt.ylabel("Total Views (in millions)", fontsize=12)
plt.xticks(rotation=45, ha="right")
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

language_counts = df["Video_language"].value_counts()

explode = [0 , 0.1]

palette = ["blue", "lightblue"]

plt.figure(figsize=(6,6))

plt.pie(language_counts, labels=language_counts.index,
        autopct="%1.1f%%", startangle=90, shadow=True,
        explode=explode, colors=palette)

plt.title("Proportion of Videos by Language", fontsize=14, fontweight="bold")
plt.legend(title="Languages", loc="best")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

channel_likes = df.groupby("Channel_name")["Likes_in_thousands"].sum().sort_values()

plt.figure(figsize=(12,10))

plt.barh(channel_likes.index, channel_likes.values, color="lightgreen",
         edgecolor="green")

plt.title("Likes of Videos by Channel",fontsize=14, fontweight="bold")
plt.xlabel("Likes (in thousands)")
plt.ylabel("Channel Name")
plt.xlim(0, 1000)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

channel_likes = df.groupby("Channel_name")["Likes_in_thousands"].sum().sort_values(ascending=False)

top_channels = channel_likes.head(15)

plt.figure(figsize=(12,8))
plt.barh(top_channels.index, top_channels.values, color="lightgreen", edgecolor="green")

plt.title("Top 15 Channels by Likes", fontsize=14, fontweight="bold")
plt.xlabel("Likes (in thousands)", fontsize=12)
plt.ylabel("Channel Name", fontsize=12)

plt.gca().invert_yaxis()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(df["Views_in_millions"], df["Likes_in_thousands"],
            color="pink", alpha=0.6, edgecolor="black")

plt.title("Scatter Plot: Views vs Likes")
plt.xlabel("Views (in millions)")
plt.ylabel("Likes (in thousands)")

plt.xlim(0, 10)
plt.ylim(0, 1000)

plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

top_videos = df.nlargest(15, "Views_in_millions")

plt.figure(figsize=(8,6))

plt.scatter(top_videos["Views_in_millions"],
            top_videos["Likes_in_thousands"],
            color="pink", alpha=0.6, edgecolor="black", s=100)

plt.title("Scatter Plot for Top 15 Videos : Views vs Likes")
plt.xlabel("Views (in millions)", fontsize=12)
plt.ylabel("Likes (in thousands)", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
sns.boxplot(x="Category", y="Views_in_millions", data=df, color="grey")

plt.xticks(rotation=300)

plt.title("Distribution of Views Across Video Categories", fontsize=16)
plt.xlabel("Category", fontsize=12)
plt.ylabel("Views (in millions)", fontsize=12)
plt.ylim(0,100)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(15,11))

plt.bar(df['Video_title'], df['Likes_in_thousands'],
        label='Likes', color='skyblue',)
plt.bar(df['Video_title'], df['Dislikes_in_thousands'],
        bottom=df['Likes_in_thousands'], label='Dislikes', color='red')

plt.xlabel("Video Title", fontsize=12)
plt.ylabel("Count (in thousands)", fontsize=12)
plt.title("Likes vs Dislikes per Video", fontsize=16)

plt.ylim(0, 1000)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df["Engagement"] = df["Likes_in_thousands"] + df["Dislikes_in_thousands"]

top_videos = df.nlargest(15, "Engagement")

plt.figure(figsize=(15,8))

plt.bar(top_videos['Video_title'], top_videos['Likes_in_thousands'],
        label='Likes', color='skyblue')
plt.bar(top_videos['Video_title'], top_videos['Dislikes_in_thousands'],
        bottom=top_videos['Likes_in_thousands'], label='Dislikes', color='red')

plt.xlabel("Video Title", fontsize=12)
plt.ylabel("Count (in thousands)", fontsize=12)
plt.title("Top 15 Videos: Likes vs Dislikes", fontsize=16)
plt.ylim(0, 1000)
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
custom_palette=["lightgreen","lightblue","green","red","blue","yellow","pink","grey"]
sns.countplot(x='Country', data=df, palette=custom_palette,
              edgecolor="black",order=df['Country'].value_counts().index)

plt.xlabel("Country", fontsize=12)
plt.ylabel("Number of Videos", fontsize=12)
plt.title("Number of Videos by Country", fontsize=16)

plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sort data by video duration
df_sorted = df.sort_values("Duration_in_minutes")

plt.figure(figsize=(10,8))

plt.plot(df_sorted["Duration_in_minutes"],
         df_sorted["Comment_in_thousands"],
         marker="o", linestyle="-", color="purple", alpha=0.5)

plt.xlabel("Video Duration (minutes)", fontsize=12)
plt.ylabel("Comments (in thousands)", fontsize=12)
plt.title("Trend of Comment Counts vs Video Duration", fontsize=16)

plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

num_cols = ["Views_in_millions", "Likes_in_thousands",
            "Dislikes_in_thousands", "Comment_in_thousands",
            "Duration_in_minutes"]

corr_matrix = df[num_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap of Engagement Metrics & Duration", fontsize=14, fontweight="bold")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))

sns.violinplot(x="Category", y="Views_in_millions",
               data=df,palette="Set2")

plt.title("Distribution of Views Across Video Categories", fontsize=16, fontweight="bold")
plt.xlabel("Video Category", fontsize=12)
plt.ylabel("Views (in millions)", fontsize=12)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,8))

sns.scatterplot(
    x="Views_in_millions", y="Likes_in_thousands",
    hue="Video_language",data=df,
    palette="Set2",s=80,
    alpha=0.7, edgecolor="black")

plt.title("Scatter Plot: Views vs Likes by Video Language", fontsize=14, fontweight="bold")
plt.xlabel("Views (in millions)", fontsize=12)
plt.ylabel("Likes (in thousands)", fontsize=12)

plt.grid(True, linestyle="--", alpha=0.5)
plt.legend(title="Language", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

sns.barplot(
    x="Category", y="Duration_in_minutes",
    data=df, estimator="mean",
    palette="Set3", edgecolor="black")

plt.title("Average Video Duration Across Categories", fontsize=16, fontweight="bold")
plt.xlabel("Video Category", fontsize=12)
plt.ylabel("Average Duration (minutes)", fontsize=12)

plt.show()
```
